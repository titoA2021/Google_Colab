```python
# Valor que maximiza 
import matplotlib.pyplot as plt
def quad(a,b,c,x):
  return a*(x**2)+b*x+c 
f = list(range(-20,21))
t= []
for valo in f:
  t.append(quad(a=-10,b=5,c=150, x=valo))
plt.plot(f, t)
print(t.index(max(t)), 'Numero que maxima es:',max(t))
```

```python
# Apartir de los elementos de una lista, volverlo una expresi√≥n numerica
#Algoritmos geneticos 
def conDec(lista):
  e = len(lista)
  acum = 0
  for i in lista:
    acum = acum + (i * (10**(e-1)))
    print(acum)
    e -= 1
  return acum

num = [3,5,2,6,7]
conDec(num)
```

```python
#factorial
def fact(n):
  acum = 1
  for i in range(1,n+1):
    acum=acum*(i)
    print(acum)
  return acum

#factorial
def fact2(n):
  acum = 1
  for i in range(n):
    acum = acum * (n - i)
    print(acum)
  return acum

fact(7)

def factW(n):
  acum = 1
  i=0
  while i <= n and (n-i) > 0:
    acum = acum * (n-i)
    print("i:", i, "n-i:", n-i)
    i = i + 1
    print("acum:", acum)
  return acum

factW(3)
```

```python
#Conjetura de collatz
import matplotlib.pyplot as plt
#Agranda la pantalla de la grafica
from matplotlib.pyplot import figure
figure(figsize=(17, 10))
def collatz(n):
  y = []
  y.append(n)
  while n > 1:
    if n%2 == 0:
      n = n/2
      y.append(n)
    else:
      n = 3*n + 1
      y.append(n)
  return y

for i in range(2,15):
  plt.plot(collatz(i))
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(14, 12))
def walk(n):
  if n==1:
    return 1
  elif n%2==0:
    return n / 2
  else:
    return 3*n + 1

def collatz(n):
  seq = []
  seq.append(n)
  while n!=1:
    n = walk(n)
    seq.append(n)
  return seq
  
for x in range(1,80):
  plt.plot(collatz(x))
```
